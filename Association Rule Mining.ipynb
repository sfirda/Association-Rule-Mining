```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
data=pd.read_csv("store_data.csv",header=None)
```

```python
data
```

#### NaN dalam data ini bukanlah Missing Data tetapi memang tidak ada transaksi yang terjadi

```python
records=[]
for i in range(0,7501): records.append([str(data.values[i,j]) for j in range (0,20) if str(data.values[i,j]) !='nan'])
```

## Applying Apriori

```python
min_purchased=35
support=min_purchased/7500
support
```

#### Note: Terdapat 4 parameter dalam menggunakan apriori, data dari record, minimum support (35) bergantung pada ketentuan dari peneliti, minimum confidence (20%), dan lift

#### Lift adalah rasio. Lift (A->B) meningkatkan penjualan B jika A dibeli. Jika Lift=1 maka tidak ada hubungan antara A dan B. Lift >1 maka jika A dibeli maka B juga dibeli. Lift<1 kedua produk tidak dibeli bersamaan.

#### min_length: kita menentuka berapa rule yang akan digunakan (2) A dan B bergantung pada peneliti

```python
association_rules=apriori(records, min_support=support, min_confidence=0.2, min_lift=3, min_length=2)
association_results=list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

#### Nilai confidence 30.07% artinya 30.07% transaksi yang mengandung mushroom sauce juga mengandung escalope. Lift>1, maka orang yang membeli mushroom juga membeli escalope.

```python
for item in association_results:
    pair=item[0]
    items=[x for x in pair]
    print("Rules: "+items[0]+"->"+items[1])
    print("Support: "+ str(item[1]))
    print("Cofindence: "+ str(item[2][0][2]))
    print("Lift: "+ str(item[2][0][3]))
    print("======================================")
```

#### Rules: Olive Oil -> Soup. Transaksi yang mengandung olive oil dari keseluruhan transaksi memiliki rasio sebesar 0.00519.  Nilai confidence 22.5% artinya 22.5% transaksi yang mengandung olive oil juga mengandung soup. lift>1 artinya maka orang yang membeli olive oil juga membeli soup.

#### Rules: Olive Oil -> Spaghetti. Transaksi yang mengandung olive oil dari keseluruhan transaksi memiliki rasio sebesar 0.00506.  Nilai confidence 20.1% artinya 20.1% transaksi yang mengandung olive oil juga mengandung spaghetti. lift>1 artinya maka orang yang membeli olive oil juga membeli sapghetti.
